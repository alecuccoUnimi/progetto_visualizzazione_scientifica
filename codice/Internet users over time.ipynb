```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 

internet = pd.read_excel("../dati/internet use around the globe.xlsx")
internet
```

```python
years = internet.iloc[[83],1:18].to_numpy()[0]
values = internet.iloc[[84],1:18].to_numpy()[0]
years = years.astype(int)
```

```python
plt.figure(figsize=(16,10),)
plt.plot(years,values, marker="o" , linewidth=5, markersize=20)
plt.grid(True)
plt.xticks(fontsize=14) 
plt.yticks(fontsize=14) 
plt.savefig('internet_users_over_time.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
