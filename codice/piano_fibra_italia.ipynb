```python
import matplotlib.pyplot as plt
import pandas as pd

years=["2022","2023","2024","2025","2026"]
values=[1,25,60,90,100]

plt.figure(figsize=(13,7))
plt.fill_between(years,0,values, color="skyblue")
plt.plot(years, values, color="Slateblue")
plt.ylabel("% Civici Coperti", fontsize=13)
plt.savefig('piano_1_giga_italia.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
