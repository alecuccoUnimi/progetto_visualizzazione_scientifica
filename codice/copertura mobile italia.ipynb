```python
import matplotlib.pyplot as plt
import pandas as pd

labels=["non coperto","coperto con velocità di picco < di 30 Mbit/s","coperto con velocità di picco >= 30 Mbit/s"]
copertura_mobile_2021 = [2.44, 24.44, 73.12]

plt.figure(figsize=(5,5))

plt.title('Copertura reti mobili 2021', fontsize=18)
plt.pie(copertura_mobile_2021, autopct='%1.2f%%', startangle=90,textprops={'fontsize': 14})
plt.legend(labels)
plt.axis('equal')
plt.savefig('copertura_mobile_2021', dpi=300, bbox_inches='tight')
plt.show()
```

```python
copertura_mobile_2026 = [2.01,13.25,84.74]
plt.figure(figsize=(5,5))
plt.title('Copertura reti Mobili 2026', fontsize=18)
_, _, autotexts = plt.pie(copertura_mobile_2026, autopct='%1.2f%%', startangle=90,textprops={'fontsize': 14})
autotexts[0]._y =+0.8
plt.legend(labels)
plt.axis('equal')
plt.savefig('copertura_mobile_2026', dpi=300, bbox_inches='tight')
plt.show()
```
