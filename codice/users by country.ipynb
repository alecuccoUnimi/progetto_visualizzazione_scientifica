```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 

internet = pd.read_excel("PercentIndividualsUsingInternet.xlsx")
list=["Denmark","United Arab Emirates","Korea (Rep. of)","United Kingdom","Sweden","Japan","France","Germany","Singapore",
      "United States","Italy","Argentina","Brazil","China","South Africa","India","Central African Rep.","Nigeria"]
internet = internet.loc[internet["Country"].isin(list)]
internet = internet.loc[:, ["Country","2020_value"]]
internet
```

```python

new_row = {"Country":"World","2020_value":62.5}
internet = internet.append(new_row, ignore_index=True)
internet
```

```python
#ordinamento valori in ordine decrescente e limito numero decimali a 2
internet = internet.sort_values(by=["2020_value"], ascending=True)
internet = internet.round(decimals=2)
internet
```

```python
plt.figure(figsize=(18,13))
barlist= plt.barh(internet['Country'], internet['2020_value'] , color="Grey")
barlist[3].set_color('coral')
barlist[6].set_color("deepskyblue")
plt.xticks(np.arange(0,110,10))
ax = plt.gca()
ax.xaxis.grid(True)
plt.xlabel("Percentuale utenti internet", fontsize=15)
plt.savefig('internet_use_by_country.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
