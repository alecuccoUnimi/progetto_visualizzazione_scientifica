```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 

internet = pd.read_csv("../dati/users_by_country_2021.csv")
list=["Denmark","United Arab Emirates","South Korea","Norway","United Kingdom","Sweden","Japan","France","Germany","Singapore",
      "United States","Italy","Argentina","Brazil","China","South Africa","India","Central African Republic","Eritrea","Nigeria"]
internet = internet.loc[internet["Unnamed: 0"].isin(list)]
internet
```

```python
#ordinamento valori in ordine decrescente e limito numero decimali a 2
internet = internet.sort_values(by=["percentuale"], ascending=True)
internet
```

```python
plt.figure(figsize=(18,13))
barlist= plt.barh(internet['Unnamed: 0'], internet['percentuale'] , color="Grey")
barlist[8].set_color("deepskyblue")
#ax = plt.gca()
#ax.xaxis.grid(True)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(np.arange(0,110,10))
ax = plt.gca()
ax.xaxis.grid(True)
plt.xlabel("Percentuale utenti internet 2021", fontsize=15)
plt.savefig('internet_use_by_country.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
