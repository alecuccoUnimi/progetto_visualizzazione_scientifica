```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

pc = pd.read_excel("uso pc italia.xlsx")
```

```python
dati2020 = pc.iloc[4:18,:]
dati2020 = dati2020.dropna(how='all', axis=1)
dati2020
```

```python
dati2021 = pc.iloc[18:33,:]
dati2021 = dati2021.dropna(how='all', axis=1)
dati2021
```

```python
ages = ["3-5","6-10","11-14","15-17","18-19","20-24","25-34","35-44","45-54","55-59","60-64","65-74","75 o pi첫"]
list1 = dati2020['Unnamed: 4'].to_numpy()
list2 = dati2021['Unnamed: 4'].to_numpy()
list = {'2020':list1[:13], '2021': list2[:13]}
df = pd.DataFrame(list, index=ages)



df.plot.bar(rot = 0, figsize=(15,11))
plt.xticks(fontsize=16) 
plt.yticks(fontsize=16) 
ax = plt.gca()
ax.yaxis.grid(True)
plt.xlabel("anni" , fontsize=15)
plt.savefig('pc_use_italy_by_age.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
